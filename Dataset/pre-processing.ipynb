```python
# convert npy files to png
import numpy as np
import cv2
import os
src_dir = r'D:\GAN\pytorch-CycleGAN-and-pix2pix\datasets\try_new\TO'
save_dir = r'D:\GAN\pytorch-CycleGAN-and-pix2pix\datasets\try_new'
dataset = os.listdir(src_dir)
os.chdir(save_dir)
for file in dataset:
    data = np.load(os.path.join(src_dir,file))
    data=np.clip(data,0,1)
    data_image=np.array(data*255, dtype=np.uint8)
    name = file[:-4]
    cv2.imwrite(name+'.png',data_image)
```

```python
## combine two images into one (invert the color of images if needed)
# -*- coding: utf-8 -*-
import os
from PIL import Image
import PIL.ImageOps 
import numpy as np
import cv2

src_dir1=  r'D:\GAN\pytorch-CycleGAN-and-pix2pix\datasets\Change of Volfrac\Combine_Volfrac_dataset\test'
src_dir2=  r'D:\GAN\pytorch-CycleGAN-and-pix2pix\datasets\Change of Volfrac\input\Chang of Volfrac_TO_test_Ori_pix2pix_predicts_epoch_epoch_295'
# src_dir1=r'datasets/GAN_Control_for_Change_of_Volfrac_Extend/testA'
# src_dir2=r'datasets/GAN_Control_for_Change_of_Volfrac_Extend/testB'
file_list=os.listdir(src_dir1)
save_dir= r'D:\GAN\pytorch-CycleGAN-and-pix2pix\datasets\Change of Volfrac\Ori_295'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
for file in file_list:
    file_path1=os.path.join(src_dir1,file)
    # file_path2=os.path.join(src_dir2,'AM_'+file.replace('_A.png','_B.png'))
    file_path2=os.path.join(src_dir2, file)   #('_A.png','_B.png')) #.replace('.png','_predict.png')
    img1 =cv2.imread(file_path1) #.convert('L')
    # Test if figure 2 exist
    if os.path.isfile(file_path2):
        img2=cv2.imread(file_path2)#.convert('L')
        # increase the width to combine two images horizontally
        img_c = cv2.hconcat([img1, img2])
        cv2.imwrite(os.path.join(save_dir,file),img_c)
        #reverse the color
        image = Image.open((os.path.join(save_dir,file)))
        inverted_image = PIL.ImageOps.invert(image)
        inverted_image.save(os.path.join(save_dir,file))
```
